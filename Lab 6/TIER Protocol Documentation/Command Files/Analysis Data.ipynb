```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('../Analysis Data/data.csv')
```

```python
data1 = data.loc[data['Płeć kupującego'] == 'M'].groupby(['Wiek kupującego']).size()
data2 = data.loc[data['Płeć kupującego'] == 'K'].groupby(['Wiek kupującego']).size()
dataToPlot = pd.concat([data1, data2], axis=1, ignore_index=True)

fig, ax = plt.subplots(figsize=(8, 5))
dataToPlot.plot.bar(ax=ax, color=['red','blue'])
ax.set_title('Ilość klientów w poszczególnych grupach wiekowych')
ax.set_xlabel('Wiek klienta')
ax.set_ylabel('Liczba klientów')
ax.legend(['Mężczyzna','Kobieta'])
plt.show()
```

Z powyższego wykresu możemy zauważyć, że rozkład ilości klientów w zależności od ich wieku jest w przybliżeniu normalny. Największa grupa klientów to ludzie w średnim wieku około 40 lat. Najmłodsze osoby są pełnoletnie, a najstarsze skończyli do 76 lat. Widać tendencję spadkową w przypadku osób starszych. Wynika to najprawdopodobniej z okresu w życiu, gdzie znajdujemy stałe miejsce zamieszkania.

```python
data1 = data.loc[data['Płeć kupującego'] == 'M'].groupby(['Marka']).size()
data2 = data.loc[data['Płeć kupującego'] == 'K'].groupby(['Marka']).size()
dataToPlot = pd.concat([data1, data2], axis=1, ignore_index=True)

fig, ax = plt.subplots(figsize=(8, 5))
dataToPlot.plot.bar(ax=ax, color=['red','blue'])
ax.set_title('Ilość klientów kupujących daną markę')
ax.set_xlabel('')
ax.set_ylabel('Liczba klientów')
ax.legend(['Mężczyzna','Kobieta'])
plt.show()
```

Z powyższego wykresu możemy zauważyć, że największy udział w sprzedaży ma marka Tefal, jednak nie jest to dominująca przewaga. Widzimy podobne różnice w porównaniu z pierwszym, a drugim itd. Najgorzej wypada firma Beko. Z wykresu możemy również odnotować, że przeważającą płcią kupującą odkurzacze są kobiety. Stosunek kobieta/mężczyzna dla każdej z marki zwiększa się z ilością sprzedanych odkurzaczy. Dla żadnej z marek nie kupiono więcej niż 20 odkurzaczy przez mężczyzn.

```python
dataToPlot = data.groupby(['Marka']).mean(numeric_only=True)['Ocena']
fig, ax = plt.subplots(figsize=(8, 5))
dataToPlot.plot.bar(ax=ax, color='blue')
ax.bar_label(ax.containers[0],fmt='%.2f')
ax.set_title('Średnia ocen dla danej marki')
ax.set_xlabel('')
ax.set_ylabel('Ocena')
plt.show()
```

Z powyższego wykresu możemy zauważyć, że średnia ocena dla poszczególnych marek jest zbliżona do siebie i zawsze przekracza wartość 2, nie przekraczają 2,5. Najlepiej ocenianie są odpowiednio marka, która sprzedała najwięcej oraz najmniej odkurzaczy. Taka informacja mówi, nam o tym, że "kupując w ciemno" prawdopodobnie najlepiej wybrać odkurzacz marki Tefal.

```python
dataToPlot = data['Dni od zakupu'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(8, 5))
dataToPlot.plot.bar(ax=ax, color='blue')
ax.set_title('Ilość ocen w zależności od czasu po zakupie')
ax.set_xlabel('Liczba dni od zakupu')
ax.set_ylabel('Ocena')
plt.show()
```

Z powyższego wykresu możemy zauważyć, że ilość ocen w zależności od czasu ma w przybliżeniu lekko przesunięty rozkład normalny. Najwięcej klientów decyduje się wystawić ocenę w okolicach tygodnia po zakupie. Jest to czas, który pozwala stwierdzić czy zakupiony produkt spełnia wymagania klienta. Z wykresu możemy stwierdzić, że jeżeli klient nie wystawi oceny w okolicach 15 dnia po zakupie, to jest małoprawdopodobne, że w ogóle ją wystawi.
