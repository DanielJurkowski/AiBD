#### Zadanie 3

```python
import numpy as np
import matplotlib.pyplot as plt

# utworzenie funkcji
def function(x):
    return x ** 2 + 5

# utworzenie wektorów
x_1 = np.linspace(-1, 1)
x_2 = np.linspace(-6, 6)
x_3 = np.linspace(0, 5)

# Sporządzenie wykresu wraz z opisami osi, tytułem oraz legendą
plt.plot(x_1, function(x_1), label="x^2 + 5")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid()
plt.title("Wykres x^2 + 5 dla x od -1 do 1")
plt.legend()
plt.show()

plt.plot(x_2, function(x_2), label="x^2 + 5")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid()
plt.title("Wykres x^2 + 5 dla x od -6 do 6")
plt.legend()
plt.show()

plt.plot(x_3, function(x_3), label="x^2 + 5")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid()
plt.title("Wykres x^2 + 5 dla x od 0 do 5")
plt.legend()
plt.show()
```
